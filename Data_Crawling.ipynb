```python
#유튜브 크롤링을 위한 모듈
from pytube import Playlist
from pytube import YouTube
from youtube_transcript_api import YouTubeTranscriptApi

import pandas as pd
```

```python
playlist = []
pl = Playlist("https://www.youtube.com/watch?v=_cR_0kXcjao&list=PLs0W8Tl4bIZC31TPEYenIrkuMtbNGO3g_")

titles = []
story = []
text = ''

for url in pl:
    key = url.split('=')
    playlist.append(key[1])

for i in range(len(playlist)):
    url = str(playlist[i])
    yt_url = YouTube('https://www.youtube.com/watch?v='+url)
    caption = yt_url.captions.get_by_language_code('a.ko')
    if caption == None:
        i+=1
    else:
        srt = YouTubeTranscriptApi.get_transcript(url, languages=['ko'])
        titles.append(yt_url.title)
        for w in range(len(srt)):
            text += srt[w]['text']+' '
        story.append(text)
        text = ''
        i+=1
```

```python
data = {'제목':titles, '동화':story}
df = pd.DataFrame(data)
df.describe()
```
